```python
import numpy as np
import matplotlib.pyplot as plt
import copy
import pandas as pd
```

# MAE from Target
### Does correction improve overall fetal fraction estimation?

```python
def plot_pre_post_correction_MAE(data, title):
    # Extract conditions, means, and standard deviations
    conditions = list(data.keys())
    means_original = [data[cond]["means"][0] for cond in conditions]
    means_corrected = [data[cond]["means"][1] for cond in conditions]
    stds_original = [data[cond]["stds"][0] for cond in conditions]
    stds_corrected = [data[cond]["stds"][1] for cond in conditions]

    # Bar width and positions
    bar_width = 0.2
    index = np.arange(len(conditions))

    # Create figure and axes
    fig, ax = plt.subplots(figsize=(5, 5))

    # Plot bars
    bars_original = ax.bar(index, means_original, bar_width, yerr=stds_original, label='original', color='red', capsize=5)
    bars_corrected = ax.bar(index + bar_width, means_corrected, bar_width, yerr=stds_corrected, label='corrected', color='green', capsize=5)

    # Add labels, title, and legend
    ax.set_xlabel('True Fetal Fraction')
    ax.set_ylabel('MAE')
    ax.set_title(title)
    ax.set_xticks(index + bar_width / 2)
    ax.set_xticklabels(conditions)
    ax.legend()

    # Show plot
    plt.tight_layout()
    # plt.savefig(f'Plots/fetal_fraction__{title}.png', dpi=300)
    plt.show()
    

def get_performance(folder, prefix, mixed_percent):
    target_fraction = mixed_percent / (100.0 + mixed_percent)
    csv_file = f'{folder}/fetal_fraction_prediction_analysis/{prefix}_fetal_fraction.csv'
    frac_arr = pd.read_csv(csv_file)['Fetal_Fraction'].values
    frac_std = round( np.std(frac_arr), 4)
    mean_dev = round( np.mean(np.abs(frac_arr - target_fraction)), 4)
    std_dev = round( np.std(np.abs(frac_arr - target_fraction)), 4)
    return mean_dev, std_dev

def visualize_pre_post_correction_MAE(is_same, coverage):
    strr = 'healthy_c2i' if is_same==True else 'mixed'
    percents = [5, 10]
    data = {}
    for percent in percents:
        folder = f'{coverage}__{strr}__{percent}_percent'
        tmp_data = {}
        original_mean, original_std = get_performance(folder, 'original', percent)
        corrected_mean, corrected_std = get_performance(folder, 'corrected', percent)
        tmp_data['means'] = [original_mean, corrected_mean]
        tmp_data['stds'] = [original_std, corrected_std]
        data[f'{percent}%'] = copy.deepcopy(tmp_data)
    sub_strr = 'Same' if is_same==True else 'Different'
    title = f'{sub_strr} Cohort {coverage} WGS'
    plot_pre_post_correction_MAE(data, title)
        
configs = [[True, '2X'], [False, '2X']]
for config in configs:
    visualize_pre_post_correction_MAE(config[0], config[1])
```

# Fragment no to Chr Len Ratio Variation Within Sample
### Does correction make this ratio more consistent across contigs of the same sample?

```python
def plot_pre_post_correction_MAE(data, title):
    # Extract conditions, means, and standard deviations
    conditions = list(data.keys())
    means_original = [data[cond][0] for cond in conditions]
    means_corrected = [data[cond][1] for cond in conditions]
    
    # Bar width and positions
    bar_width = 0.25
    index = np.arange(len(conditions))

    # Create figure and axes
    fig, ax = plt.subplots()

    # Plot bars
    bars_original = ax.bar(index, means_original, bar_width, label='original', color='red', capsize=5)
    bars_corrected = ax.bar(index + bar_width, means_corrected, bar_width, label='corrected', color='green', capsize=5)

    # Add labels, title, and legend
    ax.set_xlabel('Mother and Child Source Cohort')
    ax.set_ylabel('Standard Deviation')
    ax.set_title(title)
    ax.set_xticks(index + bar_width / 2)
    label_conditions = []
    for condition in conditions:
        cohort_name = condition.split('__')[0]
        sample_type = condition.split('__')[1]
        label_conditions.append( f'{cohort_name}\n{sample_type}' )
    ax.set_xticklabels(label_conditions)
    ax.legend()

    # Show plot
    plt.tight_layout()
    plt.savefig('Plots/fragment_to_len_ratio.png', dpi=300)
    

def get_std(csv_file):
    X = pd.read_csv(csv_file)['Fragment_to_Length ratio'].values
    X = X / np.max(X)
    return np.std(X)


sample_names = ['healthy_c2i__female_mother', 'healthy_c2i__male_baby', 'healthy_invitro__female_mother']
folder = 'sample_contig_frag_chr_len_ratio_info'
data = {}
for sample_name in sample_names:
    original_std = get_std(f'{folder}/{sample_name}__original.csv')
    corrected_std = get_std(f'{folder}/{sample_name}__corrected.csv')
    data[sample_name] = [original_std, corrected_std]
title = f' Variation in Fragment No to Contig Length Ratio Within Sample'
plot_pre_post_correction_MAE(data, title)
```
