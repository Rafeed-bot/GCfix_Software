```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
```

```python
file_path = "Data/NDR_Scores.parquet"
sample_list = pd.read_parquet(file_path, columns=['sid', 'transcript', 'ndr', 'ndr_type'], engine='pyarrow').values.tolist()
# sample, gene_id, ndr_score, 'gc_correction'

file_path = "Data/GTEx_Blood_TPM-500_sites.csv"
gene_list = pd.read_csv(file_path, usecols=['transcript', 'group']).values.tolist()
# gene_id, TPM

sample_gene_list = []
for sample in sample_list:
    sample_id, gene_id, ndr_score, ndr_type = sample[0], sample[1], sample[2], sample[3]
    for gene in gene_list:
        if gene[0]==gene_id:
            sample_gene_list.append([sample_id, gene_id, gene[1], ndr_score, ndr_type]) 
            # sample_id, gene_id, TPM, ndr_score, gc_correction_or_not
            break

df = pd.DataFrame(sample_gene_list)
df.to_csv('Data/NDR_Figure_Data.csv', index=False, header=['sample_id', 'gene_id', 'TPM', 'ndr_score', 'ndr_type'])
```

```python
df = pd.read_csv('Data/NDR_Figure_Data.csv')
all_samples = list( set( df['sample_id'].values.tolist() ) )
all_tpms = ['TPM <= 0.01', '0.01 <= TPM < 0.1', '0.1 <= TPM < 0.2', '0.2 <= TPM < 0.5', '0.5 <= TPM < 1', '1 <= TPM < 3',
           '3 <= TPM < 5', 'TPM > 5']


def make_box_plot(ndr_label):
    global all_samples, all_tpms
    tpm_box_dic = {}
    for tpm in all_tpms:
        tpm_box_dic[tpm] = []
        for sample in all_samples:
            tpm_box_dic[tpm].append( np.mean( df[(df['sample_id'] == sample) & (df['TPM'] == tpm) & (df['ndr_type'] == ndr_label)]['ndr_score'].values.tolist() ) )
    
    for i in range(len(all_tpms)-1):
        prev_tpm, post_tpm = all_tpms[i], all_tpms[i+1]
        ndrs_prev, ndrs_post = tpm_box_dic[prev_tpm], tpm_box_dic[post_tpm]
        _, p_value = mannwhitneyu(ndrs_prev, ndrs_post, alternative='greater')
        print(f'({prev_tpm}) > ({post_tpm}): {p_value}')
    
    labels = list(tpm_box_dic.keys())
    data = list(tpm_box_dic.values())
    plt.boxplot(data, labels=labels)
    plt.xticks(rotation=45, ha='right')
    plt.xlabel('Gene Expression (TPM)')
    plt.ylabel('NDR Score')
    plt.tight_layout()
    fig_name = 'Original' if ndr_label=='read' else 'Corrected'
    plt.savefig(f'Plots/{fig_name}.png', dpi=300)
```

```python
make_box_plot('read')
```

```python
make_box_plot('gc_flag')
```
