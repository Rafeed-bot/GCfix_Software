# Preprocessing

```python
def interpolate_zero_elements(array):
    # Find the start and end indices where the first and last non-zero elements occur
    first_non_zero_index = np.argmax(array != 0)
    last_non_zero_index = len(array) - 1 - np.argmax(array[::-1] != 0)
    # Isolate the portion of the array that needs interpolation
    inner_array = array[first_non_zero_index:last_non_zero_index + 1]
    # Get the indices of the non-zero elements in the inner array
    non_zero_indices = np.nonzero(inner_array)[0]
    # Get the indices of the zero elements in the inner array
    zero_indices = np.where(inner_array == 0)[0]
    # Interpolate to fill in the zero elements with the nearest non-zero values
    inner_array[zero_indices] = np.interp(zero_indices, non_zero_indices, inner_array[non_zero_indices])
    # Place the interpolated inner array back into the original array
    array[first_non_zero_index:last_non_zero_index + 1] = inner_array
    return array

def process_length_level(len_sample):
    processed_sample = np.zeros( len_sample.shape )
    for i in range(len_sample.shape[0]):
        processed_sample[i, :] = interpolate_zero_elements(len_sample[i, :])
    return processed_sample

ref_len_GC_array = process_length_level(ref_GC_array)
len_GC_array_dic = {}
for key in GC_array_dic.keys():
    len_GC_array_dic[key] = process_length_level(GC_array_dic[key])
```

# General GC Bias for Cohort Samples

```python
lower_cnt, upper_cnt = 10, 85
thr = 3
lag = 10

def remove_outlier(arr):
    high = np.percentile(arr, 100-thr)
    low = np.percentile(arr, thr)
    arr = np.where(arr>high, high, arr)
    arr = np.where(arr<low, low, arr)
    return arr 

def construct_GC_bias(sample, ref, X):
    safe_ref = ref + 1
    GC_array = sample / safe_ref
    GC_array = GC_array[lower_cnt: upper_cnt+1]
    GC_array = remove_outlier(GC_array)
    GC_array = GC_array / np.mean(GC_array)
    lowess = sm.nonparametric.lowess(GC_array, X, frac=0.1)
    GC_array = lowess[:, 1]
    GC_array = GC_array / np.mean(GC_array)
    return GC_array
    

def plot_GC_bias(sample):
    global len_GC_array_dic, ref_len_GC_array, lower_cnt, upper_cnt
    sample_GC_array = np.sum(len_GC_array_dic[sample], axis=0)
    ref_GC_array = np.sum(ref_len_GC_array, axis=0)
    X = [float(i/100) for i in range(lower_cnt, upper_cnt+1)]
    Y = construct_GC_bias(sample_GC_array, ref_GC_array, X)
    plt.plot(X[lag:-lag], Y[lag:-lag], label=sample)
    

plt.figure(figsize=(10, 6))
for key in GC_array_dic.keys():
    plot_GC_bias(key)
plt.xlabel('GC Content')
plt.ylabel('Smoothed Fragment Rate')
plt.legend()
plt.show()
```

# Fragment Grouping Analysis

```python
lower_cnt, upper_cnt = 10, 85
thr = 3
lag = 10
window_size = 5

def remove_outlier(arr):
    high = np.percentile(arr, 100-thr)
    low = np.percentile(arr, thr)
    arr = np.where(arr>high, high, arr)
    arr = np.where(arr<low, low, arr)
    return arr


def construct_GC_bias(sample, ref, X):
    safe_ref = ref + 1
    GC_array = sample / safe_ref
    
    GC_array = GC_array[lower_cnt: upper_cnt+1]
    GC_array = remove_outlier(GC_array)
    GC_array = GC_array / np.mean(GC_array)
    
    lowess = sm.nonparametric.lowess(GC_array, X, frac=0.2)
    GC_array = lowess[:, 1]
    GC_array = GC_array[lag:-lag]
    GC_array = GC_array / np.mean(GC_array)
    return GC_array


def plot_len_GC_bias(sample, frag_len):
    global len_GC_array_dic, ref_len_GC_array, lower_cnt, upper_cnt
    sample_len_GC_array = len_GC_array_dic[sample]
    X = [float(i/100) for i in range(lower_cnt, upper_cnt+1)]
    start_len = frag_len
    end_len = start_len + window_size - 1
    plt.figure(figsize=(10, 6))
    for length in range(start_len, end_len+1):
        Y = construct_GC_bias(sample_len_GC_array[length-51], ref_len_GC_array[length-51], X)
        plt.plot(X[lag:-lag], Y)
    plt.title(f'{sample}: {start_len}-{end_len}bp')
    plt.xlabel('GC Content')
    plt.ylabel('Smoothed Fragment Rate')
    plt.show()
    

def get_len_bias(frag_len):
    global GC_array_dic, cohort
    i = 0
    for key in GC_array_dic.keys():
        plot_len_GC_bias(key, frag_len)
        i += 1
        if i>=3:
            break
```

# How the Fragment Bias Curves Look for Different Cohorts

```python
lower_cnt, upper_cnt = 10, 85
thr = 3
lag = 10


len_group_dic = {51: [51, 55], 52: [51, 55], 399: [396, 400], 400: [396, 400]}
for length in range(53, 399):
    len_group_dic[length] = [length-2, length+2]


def remove_outlier(arr):
    high = np.percentile(arr, 100-thr)
    low = np.percentile(arr, thr)
    arr = np.where(arr>high, high, arr)
    arr = np.where(arr<low, low, arr)
    return arr

def construct_GC_bias(sample_lengths, ref_lengths, X):
    sample = np.sum(sample_lengths, axis=0)
    ref = np.sum(ref_lengths, axis=0)
    safe_ref = ref + 1
    GC_array = sample / safe_ref
    
    GC_array = GC_array[lower_cnt: upper_cnt+1]
    GC_array = remove_outlier(GC_array)
    GC_array = GC_array / np.mean(GC_array)
    
    lowess = sm.nonparametric.lowess(GC_array, X, frac=0.2)
    GC_array = lowess[:, 1]
    GC_array = GC_array[lag:-lag]
    GC_array = GC_array / np.mean(GC_array)
    min_positive = np.min(GC_array[GC_array > 0])
    GC_array[GC_array <= 0] = min_positive
    return GC_array


def plot_len_GC_bias(sample, color, frag_len, ax):
    global ref_len_GC_array, len_group_dic, len_GC_array_dic
    sample_len_GC_array = len_GC_array_dic[sample]
    X = [float(i/100) for i in range(lower_cnt, upper_cnt+1)]
    start_len_ind = len_group_dic[frag_len][0] - 51
    end_len_ind = len_group_dic[frag_len][1] - 51
    Y = construct_GC_bias(sample_len_GC_array[start_len_ind: end_len_ind+1], ref_len_GC_array[start_len_ind: end_len_ind+1], X)
    ax.plot(X[lag:-lag], Y, label=sample, color=color)
    

def get_len_bias(frag_len, ax):
    sample_colors = {'D7': 'green',
                     'D14': 'olive',
                     'E6c': 'red',
                     'E8c': 'orange'}
    for sample in sample_colors.keys():
        plot_len_GC_bias(sample, sample_colors[sample], frag_len, ax)
    ax.set_title(f'{frag_len}bp')
    ax.set_xlabel('GC Content')
    ax.set_ylabel('Smoothed Fragment Rate')
    ax.legend()
```

```python
lengths = [51, 71, 91, 151, 201, 251, 301, 351, 400]
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
i = 0
for length in lengths:
    row = i // 3
    col = i % 3  
    get_len_bias(length, axes[row, col])
    i += 1
plt.tight_layout()
plt.show()
```

# Visualizing the Impact of Post Processing Outlier Removal

```python
sample = 'D14'
lower_cnt, upper_cnt = 10, 85
thr = 3
lag = 10

len_group_dic = {51: [51, 55],
                 52: [51, 55],
                 399: [396, 400],
                 400: [396, 400]}
for length in range(53, 399):
    len_group_dic[length] = [length-2, length+2]
    

def remove_outlier(arr):
    high = np.percentile(arr, 100-thr)
    low = np.percentile(arr, thr)
    arr = np.where(arr>high, high, arr)
    arr = np.where(arr<low, low, arr)
    return arr

def construct_GC_bias(sample_lengths, ref_lengths, X):
    sample = np.sum(sample_lengths, axis=0)
    ref = np.sum(ref_lengths, axis=0)
    safe_ref = ref + 1
    GC_array = sample / safe_ref
    
    GC_array = GC_array[lower_cnt: upper_cnt+1]
    GC_array = remove_outlier(GC_array)
    GC_array = GC_array / np.mean(GC_array)
    
    lowess = sm.nonparametric.lowess(GC_array, X, frac=0.2)
    GC_array = lowess[:, 1]
    GC_array = GC_array / np.mean(GC_array)
    return GC_array


def plot_len_GC_bias(frag_len, ax):
    global ref_len_GC_array, len_group_dic 
    sample_len_GC_array = len_GC_array_dic[sample]
    X = [float(i/100) for i in range(lower_cnt, upper_cnt+1)]
    start_len_ind = len_group_dic[frag_len][0] - 51
    end_len_ind = len_group_dic[frag_len][1] - 51
    Y1 = construct_GC_bias(sample_len_GC_array[start_len_ind: end_len_ind+1], 
                          ref_len_GC_array[start_len_ind: end_len_ind+1], X)
    ax.plot(X[lag:-lag], Y1[lag:-lag], label='Normal_Bias')
    
    Y2 = Y1[lag: -lag]
    min_thr = min(1/20, np.percentile(Y2, 1))
    Y2[Y2<min_thr] = min_thr
    ax.plot(X[lag:-lag], Y2, label='Bias_Min_Thresholding')

    
    
lengths = [51, 71, 91,
           150, 200, 250,
           300, 350, 400]
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
for ax in axes.flat:
    ax.set_aspect('auto') 
    
i = 0
for length in lengths:
    row = i // 3
    col = i % 3
    ax = axes[row, col]
    plot_len_GC_bias(length, ax)
    ax.set_title(f'{sample}: {length}bp')
    ax.set_xlabel('GC Content')
    ax.set_ylabel('Smoothed Fragment Rate')
    ax.legend()
    i += 1
plt.tight_layout()
plt.show()
```
